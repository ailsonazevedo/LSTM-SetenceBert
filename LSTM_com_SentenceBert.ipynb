```python
import keras
(x_train, y_train), (x_test, y_test) = keras.datasets.imdb.load_data(num_words=3000)
print("Tamanho do treino:", x_train.shape)
print("Tamanho do teste:", x_test.shape)
```

```python
qt = 3000
x_train = x_train[:qt]
y_train = y_train[:qt]
x_test = x_test[:qt]
y_test = y_test[:qt]
```

```python
def convert_to_text(dataset, words):
    aux_matrix = dict((i, word) for (word, i) in words.items())
    sentences = []
    for x in dataset:
        texto_original = " ".join(aux_matrix[i] for i in x)
        sentences.append(texto_original)
    return sentences
```

```python
words_index = keras.datasets.imdb.get_word_index()
sentence_x = convert_to_text(x_train, words_index)
print(x_train[0])
print(sentence_x[0])
```

```python
x_train_emb = model.encode(sentence_x)
```

```python
x_train_emb.shape
max_features = x_train_emb.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense, Embedding

lstm = Sequential()
lstm.add(Embedding(max_features, 364))
lstm.add(LSTM(256, dropout=0.2, recurrent_dropout=0.2))
lstm.add(Dense(1, activation='sigmoid'))

lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics = ['accuracy'])
lstm.summary()
```

```python
lstm.fit(x_train_emb, y_train, epochs = 10, batch_size=10)
```
